```python
import numpy as np
import scipy.linalg as sla
import matplotlib.pyplot as plt
```

Строим вектор из покоординатно независимых случайных величин, таких что в $k$-ой координате $\mathcal{N}(k, k)$. Отобразим истинную и сэмплированную плотность на $[-15, 15]$

```python
def normal_dens(v, mu, sigma):
    return np.exp(-(v - mu)**2/(2 * sigma)) / (np.sqrt(2 * np.pi) * np.sqrt(sigma))

for k in range(10):
    tens = np.load(f'normal_density_{k}.npy')
    plt.plot(tens[:, 0], tens[:, 1], label='marginal density')
    plt.plot(tens[:, 0], normal_dens(tens[:, 0], k+1, k+1), label='true density')
    plt.title(f'mu={k+1} sigma={k+1}')
    plt.legend()
    plt.show()
```

Строим вектор из покоординатно независимых величин, таких что $k$-ая координата это $\text{Exp}(k)$. Носитель $[0, 6]$.

```python
def exp_dens(v, la):
    v = v.copy()
    u = v[v>0]
    u = la * np.exp(-la* u)
    v[v>0] = u
    v[v <= 0] = 0.
    return v

for k in range(10):
    tens = np.load(f'exp_density_{k}.npy')
    plt.title(f'la={k+1}')
    plt.plot(tens[:, 0], tens[:, 1], label='marginal density')
    plt.plot(tens[:, 0], exp_dens(tens[:, 0], k+1), label='true density')
    plt.legend()
    plt.show()
    
```
